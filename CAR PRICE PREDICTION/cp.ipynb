```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
a = pd.read_csv("car data.csv")
```

```python
a.info()
```

```python
a.isnull().sum()
```

```python
a.describe()
```

```python
a.head()
```

```python
print(a['Fuel_Type'].value_counts())
print(a['Selling_type'].value_counts())
print(a['Transmission'].value_counts())
```

```python
a.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
```

```python
a.replace({'Selling_type':{'Dealer':0,'Individual':1}},inplace=True)
```

```python
a.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
a.head()
```

```python
X = a.drop(['Car_Name','Selling_Price'],axis=1)
```

```python
Y = a['Selling_Price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```
